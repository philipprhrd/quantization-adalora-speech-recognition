# Common Voice

```python
from datasets import Dataset, Audio, Features, Value
import pandas as pd
from pathlib import Path
import os
```

```python
def create_common_voice_dataset(path: Path) -> Dataset:
    df = pd.read_csv(path / "validated.tsv", sep="\t")

    df["audio"] = df["path"].apply(lambda x: os.path.join(path, "clips", x))

    df = df[["audio", "sentence"]]

    features = Features({
        "audio": Value("string"),
        "sentence": Value("string"),
    })

    ds = Dataset.from_pandas(df, features=features)
    ds = ds.cast_column("audio", Audio())

    return ds

ds = create_common_voice_dataset(Path('/Volumes/Philipps T7/common_voice/cv-corpus-24.0-2025-12-05/de'))

ds[0]
```
